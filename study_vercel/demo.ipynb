```typescript
import * as dotenv from "dotenv"
```

```typescript
dotenv.config({ path: "../.env" });

console.log(process.env.OPENAI_API_KEY);
console.log(process.env.OPENAI_BASE_URL);
console.log(process.env.OPENAI_MODEL);
```

```typescript
import { createOpenAICompatible } from "@ai-sdk/openai-compatible";
```

```typescript
// 配置供应商
const provider = createOpenAICompatible({
    name: "openai-compatible",
    baseURL: process.env.OPENAI_BASE_URL as string,
    apiKey: process.env.OPENAI_API_KEY as string,
});
```

```typescript
const model = provider(process.env.OPENAI_MODEL);
```

```typescript
// 生成对话
import { generateText,streamText, type ModelMessage, type UserModelMessage } from "ai";
```

```typescript
const result = await generateText({
    model: model,
    system: "Mimo operates in Concise Mode by minimizing output tokens through the elimination of unnecessary preamble, postamble, tangential details, excessive pleasantries, redundant offers, and overly comprehensive lists while preserving full helpfulness, accuracy, completeness, quality, evidence, specific data, and identical standards for code/artifacts/outputs—only expanding responses or disclosing the mode if users explicitly request longer answers or show frustration.",
    prompt: "Can you tell me the meaning of love in one sentence?",
})

console.log(result.text)
```

```typescript
import {stdout} from "node:process"
```

```typescript
const result =  streamText({
    model: model,
    prompt: "hi",
});

let full_text:string = "";

for await(const textPart of result.textStream) {
    full_text += textPart;
};

console.log(full_text);
```

```typescript
let messages: ModelMessage[] = [];
```

```typescript
messages.push({role:"user", content:"hi, My name is Zane"} as UserModelMessage);
```

```typescript
console.log(messages);
```

```typescript
// 带有状态的回复
const result1 = await generateText({
            model,
            system: "Mimo operates in Concise Mode by minimizing output tokens through the elimination of unnecessary preamble, postamble, tangential details, excessive pleasantries, redundant offers, and overly comprehensive lists while preserving full helpfulness, accuracy, completeness, quality, evidence, specific data, and identical standards for code/artifacts/outputs—only expanding responses or disclosing the mode if users explicitly request longer answers or show frustration.",
            messages: messages,
        });

messages.push(...result1.response.messages)
messages.push({role:"user", content:"Can you tell me what my name is"} as UserModelMessage);
const result2 = await generateText({
            model,
            system: "Mimo operates in Concise Mode by minimizing output tokens through the elimination of unnecessary preamble, postamble, tangential details, excessive pleasantries, redundant offers, and overly comprehensive lists while preserving full helpfulness, accuracy, completeness, quality, evidence, specific data, and identical standards for code/artifacts/outputs—only expanding responses or disclosing the mode if users explicitly request longer answers or show frustration.",
            messages: messages,
        });
messages.push(...result2.response.messages)
console.log(result2.text);
```
